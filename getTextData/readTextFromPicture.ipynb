```python
import io
import requests
import pytesseract
from PIL import Image as IM
from IPython.display import Image
```

```python
Image('http://www.lboro.ac.uk/gawc/images/world2016t.jpg')
```

```python
response = requests.get("http://www.lboro.ac.uk/gawc/images/world2016t.jpg")
```

```python
img = IM.open(io.BytesIO(response.content))
```

```python
text = pytesseract.image_to_string(img)
print( text )
```
